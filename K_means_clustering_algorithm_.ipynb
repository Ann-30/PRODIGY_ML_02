```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/Mall_Customers.csv')

# Select features for clustering, ensuring correct column names:
# ---CHANGE--- Convert 'Gender' to numerical using one-hot encoding
X = pd.get_dummies(df[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']], columns=['Gender'], drop_first=True)
# drop_first=True to avoid multicollinearity

# Initialize KMeans
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
kmeans.fit(X)

# Predict the clusters
df['Cluster'] = kmeans.predict(X)

# Visualize the clusters
sns.pairplot(df, hue='Cluster', palette='viridis', diag_kind='kde')
plt.show()

# Print the cluster centers
print('Cluster Centers:')
print(kmeans.cluster_centers_)
```
